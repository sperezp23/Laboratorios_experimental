## Histograma y Medidas estadísticas

```python
import numpy as np
import pandas as pd
import statistics
import matplotlib.pyplot as plt
import seaborn as sns

datos_df  = pd.read_excel(r'Datos\Parcial_2.xlsx', header=0)

media_tiempo = statistics.mean(datos_df.tiempo_de_oscuridad_s)
mediana_tiempo = statistics.median(datos_df.tiempo_de_oscuridad_s)
moda_tiempo = statistics.mode(datos_df.tiempo_de_oscuridad_s)
std_tiempo = np.std(datos_df.tiempo_de_oscuridad_s)

print(f'''
RESPUESTAS\n\nMedia: {media_tiempo : .2g}
Mediana: {mediana_tiempo : .2g}
Moda: {moda_tiempo : .2g}
Desviación estandar muestral: {std_tiempo : .2g}
Error estandar: {std_tiempo/6 : .2g}
\n''')

sns.histplot(datos_df.tiempo_de_oscuridad_s, bins= 6)
plt.show()
```

## Tabla de frecuencias  

```python
datos_df['intervalos'] = pd.cut(datos_df.tiempo_de_oscuridad_s, bins = 6, right=False)
tabla_de_frecuencias = datos_df[['tiempo_de_oscuridad_s', 'intervalos']].groupby('intervalos',  observed=False).count()
tabla_de_frecuencias.rename(columns={'tiempo_de_oscuridad_s':'frecuencia_absoluta'}, inplace = True)
tabla_de_frecuencias
```

## Velocidad promedio

```python
velocidad_promedio_m_s = datos_df.velocidad_m_s.mean()
error_efectivo = np.sqrt(datos_df.error_velocidad_m_s.max()**2 + (datos_df.velocidad_m_s.std()/6)**2)

print(f'RESPUESTAS:\n\nPromedio: {velocidad_promedio_m_s: .2g} [m/s]\nError efectivo:{error_efectivo : .2g} [m/s]')
print(f'Velocidad en formato científico: (420 ± 83)x10^-4 [m/s]')
```
